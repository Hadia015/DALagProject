```python
#Importing pandas
import pandas as pd
#Importing CSV
df=pd.read_csv('https://raw.githubusercontent.com/CunyLaguardiaDataAnalytics/datasets/master/2014-15_To_2016-17_School-_Level_NYC_Regents_Report_For_All_Variables.csv')
```

```python
#picking a school as an initial comparison point
school1= "Harriet Tubman Charter School"
school1_data=df[df['School Name']==school1]
```

```python
#Descriptive statistics
school1_data.describe()
```

```python
#picking a feature/measure/score "Total Tested"
print(school1_data['Total Tested'].describe())
```

```python
#Compare above school with another school
school2= "Icahn Charter School"
school2_data=df[df['School Name']==school2]
```

```python
#Descriptive statistics
school2_data.describe()
```

```python
#comparing school1 and school2 for Total Tested
print(school1_data['Total Tested'].describe())
print(school2_data['Total Tested'].describe())
```

```python
#one cleaning task renaming column
df.rename(columns={'Regents Exam': 'Subject'}, inplace=True)
```

```python
df.head()
```

```python
#one visualization
#importing matplotlib
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots()
plt.bar(school1_data['School Name'], school1_data['Total Tested'])
plt.bar(school2_data['School Name'], school2_data['Total Tested'])
ax.legend([school1, school2])
ax.set_xlabel('School Name')
ax.set_ylabel('Total Tested')
ax.set_title('Comparison of Total Tested by School')
plt.show()
```
