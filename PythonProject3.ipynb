```python
#cleaning data task
#deleting columns
drop_columns = ['X_COORD_CD','Y_COORD_CD', 'Latitude', 'Longitude', 'New Georeferenced Column']
df.drop(drop_columns, axis=1, inplace=True)
```

```python
df.columns
```

```python
#cleaning task
#assigning borough names
df['ARREST_BORO']=df['ARREST_BORO'].replace({'K': 'Brooklyn', 'B': 'Bronx', 'M': 'Manhattan', 'Q': 'Queens', 'S': 'Staten Island'})
```

```python
print(df['ARREST_BORO'])
```

```python
#pick any zip code/neighborhood/district- Manhattan
manhattan_df = df[df['ARREST_BORO'] == 'Manhattan']
print(manhattan_df)
```

```python
#descriptive statistics
manhattan_df.describe()
```

```python
#pick a feature/measure/column offense description
manhattan_df['PERP_SEX'].value_counts()
```

```python
#compare to any other entity- Brooklyn
brooklyn_df = df[df['ARREST_BORO'] == 'Brooklyn']
print(brooklyn_df)
```

```python
#descriptive statistics
brooklyn_df.describe()
```

```python
brooklyn_df['PERP_SEX'].value_counts()
```

```python
#plot manhattan value counts vs brooklyn value counts
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#plotting gender for brooklyn and manhattan
plt.plot(manhattan_df['PERP_SEX'].value_counts(), label='Manhattan')
plt.plot(brooklyn_df['PERP_SEX'].value_counts(), label='Brooklyn')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Gender Comparison')
plt.legend()
plt.show()
```

```python
#conclusion- When comparing Manhattan and Brooklyn arrests, Brooklyn had more arrests than Manhattan. This stayed consistent when comparing genders as Male and Female arrest count was higher for Brooklyn than it was for Manhattan.
```
